```python
data2 = pd.read_excel("./data03.xlsx")
data2.head()
```

```python
data2['birthday'] = pd.to_datetime(data2['birthday'],format="%Y/%m/%d")
# data2['isleap'] = data2['birthday'].dt.is_leap_year
data2['birthday'].dt.quarter
```

```python
data2.groupby(by='gender').min()
```

```python
import numpy as np
data2.agg({'tel':'max','income':np.mean},axis=0)
```

```python
def isleap(row):
    if(row['birthday'].is_leap_year):
        return "闰年"
    else:
        return "不是闰年"
data2['闰年判断'] = data2.apply(isleap,axis=1)
```

```python
def income(row):
    if(row['income']>12000):
        return "高收入"
    else:
        return "普通收入"
data2['收入判断'] = data2.apply(income,axis=1)
```
